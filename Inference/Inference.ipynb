```python
from ultralytics import YOLO
import cv2
import math 
# start webcam
cap = cv2.VideoCapture(0) # this is our only attached webcam
cap.set(3, 1920) # webcam horizontal resolution
cap.set(4, 1080) # webcam vertical resolution

# model with trained weights
model = YOLO("best.pt")


# object classes
classNames = ["Medic Veteran", "Gunner Veteran", "Flamer Veteran", "Grenade Launcher Veteran", 
              "Hardened Veteran", "Comms Veteran", "Sniper Veteran", "Spotter Veteran", 
              "Confident Veteran", "Kommando Nob", "Kommando Comms", "Kommando Slasha", 
              "Kommando Rokkit", "Kommando Slugga", "Kommando Dakka", "Kommando Choppa", 
              "Kommando Burna", "Kommando Snipa"]


while True:
    success, img = cap.read()
    results = model(img, stream=True)

    # coordinates
    for r in results:
        boxes = r.boxes

        for box in boxes:
            # bounding box
            x1, y1, x2, y2 = box.xyxy[0]
            x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2) # convert to int values

            # put box in cam
            cv2.rectangle(img, (x1, y1), (x2, y2), (255, 0, 255), 3)

            # confidence level                        
            confidence = math.ceil((box.conf[0]*100))/100
            if confidence > 0.4: # this can be adjusted to fine tune the bounding box class selection
                print("Confidence --->",confidence)

                # class name
                cls = int(box.cls[0])
                print("Class name -->", classNames[cls])

                # object details
                org = [x1, y1]
                font = cv2.FONT_HERSHEY_SIMPLEX
                fontScale = 1
                color = (255, 0, 0)
                thickness = 2

                cv2.putText(img, classNames[cls], org, font, fontScale, color, thickness)

    cv2.imshow('Webcam', img)
    if cv2.waitKey(1) == ord('q'): # use 'q' to end the inference session
        break

cap.release()
cv2.destroyAllWindows()
```
